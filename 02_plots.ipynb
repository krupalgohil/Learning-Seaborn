```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
penguins = sns.load_dataset("penguins")

# ---------------------- 1. HISTOGRAM ----------------------
sns.histplot(data=penguins, x="body_mass_g", hue="sex", multiple="stack", bins=20, kde=True)
plt.title("Histogram: Distribution of Body Mass by Sex")
plt.show()
print(" Histogram: Shows how body mass is distributed across sexes. Helps detect normality, skewness, and outliers.")

# ---------------------- 2. REGRESSION PLOT ----------------------
sns.regplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
plt.title("Regression Plot: Body Mass vs Flipper Length")
plt.show()
print(" Regression Plot: Shows linear trend between body mass and flipper length. We can check strength and slope of correlation.")

# ---------------------- 3. LINE PLOT ----------------------
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="sex", style="island", markers=True)
plt.title("Line Plot: Body Mass vs Flipper Length (by Sex & Island)")
plt.show()
print(" Line Plot: Useful for trends. Here, it shows group-wise trend lines of flipper length with respect to body mass.")

# ---------------------- 4. JOINT PLOT ----------------------
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="sex", kind="scatter")
plt.suptitle("Joint Plot: Body Mass vs Flipper Length", y=1.02)
plt.show()
print(" Joint Plot: Combines scatterplot and histograms. Helps us see distribution and correlation at the same time.")

# ---------------------- 5. BAR PLOT ----------------------
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", ci=None)
plt.title("Bar Plot: Average Body Mass by Species and Sex")
plt.show()
print(" Bar Plot: Shows which species and gender have the highest or lowest average body mass.")

# ---------------------- 6. COUNT PLOT ----------------------
sns.countplot(data=penguins, x="species", hue="sex")
plt.title("Count Plot: Number of Penguins by Species and Sex")
plt.show()
print(" Count Plot: Shows frequency of penguins in each species. Helps detect class imbalance.")

# ---------------------- 7. BOX PLOT ----------------------
sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex")
plt.title("Box Plot: Body Mass Distribution by Species and Sex")
plt.show()
print(" Box Plot: Shows spread, median, and outliers of body mass across groups.")

# ---------------------- 8. VIOLIN PLOT ----------------------
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split=True)
plt.title("Violin Plot: Body Mass Distribution Shape")
plt.show()
print(" Violin Plot: Shows both summary statistics and density distribution shape (normal/skewed/bimodal).")

# ---------------------- 9. KDE PLOT ----------------------
sns.kdeplot(data=penguins, x="body_mass_g", hue="species", fill=True)
plt.title("KDE Plot: Density of Body Mass by Species")
plt.show()
print(" KDE Plot: Shows smooth distribution of body mass. Helps visualize overlap and differences between species.")

# ---------------------- 10. HEATMAP ----------------------
corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap: Correlation Matrix of Numeric Features")
plt.show()
print(" Heatmap: Shows correlations. Strong positive/negative relationships can be detected (e.g., body_mass vs flipper_length).")

# ---------------------- 11. RUG PLOT ----------------------
sns.rugplot(data=penguins, x="body_mass_g", hue="species")
plt.title("Rug Plot: Individual Data Points of Body Mass")
plt.show()
print(" Rug Plot: Shows raw data distribution along an axis. Useful for spotting clustering.")

# ---------------------- 12. PAIR PLOT ----------------------
sns.pairplot(penguins, hue="species", diag_kind="kde")
plt.suptitle("Pair Plot: Relationships Between Numeric Variables", y=1.02)
plt.show()
print(" Pair Plot: Shows scatter plots between all numeric features + distribution. Helps detect pairwise relationships and clusters.")

# ---------------------- 13. PAIR GRID ----------------------
g = sns.PairGrid(penguins, hue="species")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot)
g.add_legend()
plt.suptitle("PairGrid: Customized Pairwise Plots", y=1.02)
plt.show()
print(" PairGrid: Like PairPlot but customizable. Allows different plots on upper, lower, and diagonal parts.")

sns.rugplot()

sns.pairplot()


sns.PairGrid()
```
