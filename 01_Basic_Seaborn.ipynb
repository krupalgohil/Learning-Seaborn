```python
'''
Step 1: Introduction to Seaborn

What is Seaborn?
Seaborn is a Python library for data visualization built on top of Matplotlib. 
It makes plots more attractive and easy to understand.

'''

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
print(tips.head())
```

```python
#Basic plot

sns.histplot(data=tips,x="total_bill",bins=20,kde=True)
plt.show()


sns.barplot(data=tips,x="day",y="total_bill")
plt.show()

sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.show()
```

```python
# Styling & Customization
# Add colors
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time")

# Change style/theme
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.set_palette("bright")
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time")
plt.show()
```

```python
# Intermediate Plots
# 1.box plot

sns.boxenplot(data=tips,x="day",y="total_bill",hue="sex")
plt.show()

#Violin plot
sns.violinplot(data=tips,x="day",y="total_bill",hue="sex",split=True)
plt.show()

#pair plot
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
# Advanced Visualization
corr = tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()


sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()

g = sns.FacetGrid(tips, col="sex", row="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()


```

## 📊 Seaborn Most Used Plots – Theory + Conclusion

---

## 📘 1. Histogram
**Theory:**  
- Shows distribution of one numeric variable.  
- Helps identify shape of data (normal, skewed, peaks).  
- Useful for spotting spread and outliers.  

**Conclusion we can take:**  
- Whether data is normally distributed.  
- Presence of skewness or outliers.  

---

## 📘 2. Scatter Plot
**Theory:**  
- Shows relationship between two numeric variables.  
- Can use `hue`, `style`, `size` for extra dimensions.  

**Conclusion:**  
- Detects correlation (positive, negative, none).  
- Finds clusters and outliers.  

---

## 📘 3. Bar Plot
**Theory:**  
- Compares average value of numeric data across categories.  
- Each bar = mean of values.  

**Conclusion:**  
- Which group has highest/lowest average.  
- Differences between categories.  

---

## 📘 4. Count Plot
**Theory:**  
- Shows frequency of categories.  
- Similar to bar plot but counts rows instead of averaging.  

**Conclusion:**  
- Which category is most common.  
- Detects imbalance in dataset.  

---

## 📘 5. Box Plot
**Theory:**  
- Displays five-number summary (min, Q1, median, Q3, max).  
- Highlights outliers separately.  

**Conclusion:**  
- Spread of data and skewness.  
- Median value of groups.  
- Presence of outliers.  

---

## 📘 6. Violin Plot
**Theory:**  
- Combination of boxplot + KDE density.  
- Shows distribution shape along with summary statistics.  

**Conclusion:**  
- Shape (normal, skewed, bimodal).  
- Compare distributions across groups.  

---

## 📘 7. Heatmap
**Theory:**  
- Represents matrix values as colors.  
- Mostly used for correlation matrices.  

**Conclusion:**  
- Strong positive/negative correlations.  
- Identify redundant variables.  

---

## 📘 8. Pair Plot
**Theory:**  
- Creates scatterplots for every pair of numeric variables.  
- Histograms on diagonals.  

**Conclusion:**  
- Overall correlation patterns.  
- Pairwise relationships and clusters.  

---

## 📘 9. Regression Plot
**Theory:**  
- Scatter plot with regression line.  
- Shows linear trend and confidence interval.  

**Conclusion:**  
- Relationship strength (strong/weak).  
- Positive/negative slope.  

---

## 📘 10. Facet Grid
**Theory:**  
- Creates multiple subplots split by categories.  
- Same plot type applied to subsets.  

**Conclusion:**  
- Compare relationships across subgroups.  
- Useful for multi-category data.
